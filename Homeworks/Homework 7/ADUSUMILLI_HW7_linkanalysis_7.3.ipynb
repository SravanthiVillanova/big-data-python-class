#### Grouping columns to In, Out Links

```python
grped_urls=blog_data[['Fromurl','Tourl']].groupby(['Fromurl'],as_index=False).count()
```

##### Initiating empty matrix of size(grped_urls)

##### Creating stochastic matrix

```python
import random
from random import randint
for i in range (0,919):
    n = grped_urls.at[i,'Tourl']
    cols = list(random.sample(range(0,919), n))
    x = 1/float(grped_urls.loc[[i]]['Tourl'])
    for j in cols:
        H[i][j]=x
```

#### Applying pagerank algorithm on matrix 'H' created above

```python
from numpy import *
 
def pagerank(H):
    n= len(H)
    w = zeros(n)
    rho = 1./n * ones(n);
    for i in range(n):
      if multiply.reduce(H[i]== zeros(n)):
        w[i] = 1
    newH = H + outer((1./n * w),ones(n))
 
    theta=0.85
    G = (theta * newH) + ((1-theta) * outer(1./n * ones(n), ones(n)))
    print (rho)
    for j in range(10):
        rho = dot(rho,G)
        print (rho)
        if j==9:
            return rho
```

```python
result = pagerank(H)
```

##### Top 10 urls as per the page rank algorithm

```python
pd.set_option('display.max_colwidth', -1)
grped_urls.loc[np.argsort(result)[-10:]]['Fromurl']
```
